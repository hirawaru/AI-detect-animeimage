# Training Results Analysis

Phân tích và trực quan hóa kết quả huấn luyện model

```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from pathlib import Path

# Setup
%matplotlib inline
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load Training History

```python
# Load training history
try:
    with open('results/training_history.json', 'r') as f:
        history = json.load(f)
    print("Training history loaded successfully")
    print(f"Total epochs: {len(history['train_loss'])}")
except Exception as e:
    print(f"Error: {e}")
    history = None
```

## 2. Plot Training & Validation Loss

```python
if history:
    fig, ax = plt.subplots(figsize=(12, 5))
    
    epochs = range(1, len(history['train_loss']) + 1)
    ax.plot(epochs, history['train_loss'], label='Train Loss', marker='o', markersize=3)
    ax.plot(epochs, history['val_loss'], label='Val Loss', marker='s', markersize=3)
    
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    ax.set_title('Training & Validation Loss')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('results/training_loss.png', dpi=150, bbox_inches='tight')
    plt.show()
```

## 3. Plot Training & Validation Accuracy

```python
if history:
    fig, ax = plt.subplots(figsize=(12, 5))
    
    epochs = range(1, len(history['train_acc']) + 1)
    ax.plot(epochs, history['train_acc'], label='Train Accuracy', marker='o', markersize=3)
    ax.plot(epochs, history['val_acc'], label='Val Accuracy', marker='s', markersize=3)
    
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Accuracy (%)')
    ax.set_title('Training & Validation Accuracy')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('results/training_accuracy.png', dpi=150, bbox_inches='tight')
    plt.show()
```

## 4. Load & Display Evaluation Metrics

```python
try:
    with open('results/metrics.json', 'r') as f:
        metrics = json.load(f)
    
    print("=== Test Set Metrics ===")
    print(f"Accuracy: {metrics['accuracy']:.4f}")
    print(f"F1-Score: {metrics['f1_score']:.4f}")
    print(f"ROC-AUC: {metrics['roc_auc']:.4f}")
    
    print("\n=== Classification Report ===")
    report = metrics['classification_report']
    for class_name in ['0', '1']:
        if class_name in report:
            print(f"\nClass {class_name}:")
            print(f"  Precision: {report[class_name]['precision']:.4f}")
            print(f"  Recall: {report[class_name]['recall']:.4f}")
            print(f"  F1-Score: {report[class_name]['f1-score']:.4f}")
except Exception as e:
    print(f"Error loading metrics: {e}")
    metrics = None
```

## 5. Display Confusion Matrix

```python
if metrics:
    cm_path = Path('results/confusion_matrix.png')
    if cm_path.exists():
        img = Image.open(cm_path)
        plt.figure(figsize=(8, 6))
        plt.imshow(img)
        plt.axis('off')
        plt.title('Confusion Matrix')
        plt.tight_layout()
        plt.show()
    else:
        print("Confusion matrix image not found")
```

## 6. Display ROC Curve

```python
if metrics:
    roc_path = Path('results/roc_curve.png')
    if roc_path.exists():
        img = Image.open(roc_path)
        plt.figure(figsize=(8, 6))
        plt.imshow(img)
        plt.axis('off')
        plt.title('ROC Curve')
        plt.tight_layout()
        plt.show()
    else:
        print("ROC curve image not found")
```

## 7. Display Precision-Recall Curve

```python
if metrics:
    pr_path = Path('results/pr_curve.png')
    if pr_path.exists():
        img = Image.open(pr_path)
        plt.figure(figsize=(8, 6))
        plt.imshow(img)
        plt.axis('off')
        plt.title('Precision-Recall Curve')
        plt.tight_layout()
        plt.show()
    else:
        print("PR curve image not found")
```

## 8. Summary Metrics Table

```python
if metrics:
    report = metrics['classification_report']
    
    data = []
    for class_name in ['0', '1']:
        if class_name in report:
            data.append({
                'Class': 'Natural' if class_name == '0' else 'Synthetic',
                'Precision': f"{report[class_name]['precision']:.4f}",
                'Recall': f"{report[class_name]['recall']:.4f}",
                'F1-Score': f"{report[class_name]['f1-score']:.4f}",
                'Support': int(report[class_name]['support'])
            })
    
    df = pd.DataFrame(data)
    print(df.to_string(index=False))
```
